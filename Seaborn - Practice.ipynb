```python
Q1 = df['tip'].quantile(0.25)
Q3 = df['tip'].quantile(0.75)
iqr = Q3 - Q1

filter = (df['tip'] >= Q1 - 1.5*iqr) & (df['tip']<=Q3 + 1.5*iqr)
```

```python
hey = df.loc[filter]
```

```python
sns.boxplot('tip', data = hey)
```
