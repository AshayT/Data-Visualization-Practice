```python
g = np.arange(0,20, 2)
h = g*3
i = g*5
plt.plot(g, h, color = 'b')
plt.plot(g, i, color = 'r')
plt.title('two line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.plot()
```

```python
plt.subplot(1,2,1) #1,2 
plt.title('two line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(g, h, color = 'g', linestyle ='-')
plt.subplot(1,2,2) #1,2
plt.plot(g, i, color = 'y', linestyle =':')
plt.show()
```

```python
plt.subplot(2,1,1)  # 2, 1
plt.title('two line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(g, h, color = 'g', linestyle ='-')
plt.subplot(2,1,2) #2, 1
plt.plot(g, i, color = 'y', linestyle =':')
plt.show()
```

```python
stud_marks = {'abhi':43, 'ritik':67, 'bob':87, 'ash':96}
names = list(stud_marks.keys())
marks = list(stud_marks.values())
plt.title('Marks of students')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.bar(names, marks)
```

```python
plt.title('Marks of students')
plt.xlabel('marks')
plt.ylabel('names')
plt.barh(names, marks, color = 'g')
```

```python
u = np.random.randint(20, 50, 15)
l = np.random.randint(10, 30, 15)
plt.scatter(u, l)
plt.show()
```

```python
u = np.random.randint(20, 50, 15)
l = np.random.randint(10, 30, 15)
v = np.random.randint(50, 100, 15)
plt.scatter(u, l, marker = '*', c = 'r', s = 50)
plt.scatter(u, v, c = 'g', s = 30)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(u, l, marker = '*', c = 'r', s = 50)
plt.subplot(1,2,2)
plt.scatter(u, v, c = 'g', s = 30)
plt.show()
```

```python
nums= [34,56,78,102,67]
fruits = ['mango', 'apple', 'orange', 'banana', 'peach']
plt.pie(nums, labels =fruits)
plt.show()
```

```python
plt.pie(nums, labels =fruits, autopct = '%0.1f%%', colors = ['blue', 'orange', 'purple', 'red', 'yellow'])
plt.show()
```
